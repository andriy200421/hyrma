Ось ваш код:

l = [1, "2", 3]

Давайте подивимося на кожен рядок окремо:

l = [1, "2", 3]: В цьому рядку ви створюєте список l з трьома елементами: цілочисельним числом 1, рядком "2" і цілочисельним числом 3.

l.pop(-2): Тут ви використовуєте метод pop(), щоб видалити другий елемент зі списку l. Однак l.pop(-2) повертає видалений елемент, тобто рядок "2".

list(l.pop(-2)): Ви намагаєтеся перетворити рядок "2" (повернутий результатом попереднього виразу) в список. Це призводить до створення списку, який містить окремі символи з рядка, тобто ['2'].

.append("тепер це останній елемент"): Ви намагаєтеся викликати метод append() на створеному списку в пункті 3. Однак метод append() модифікує список на місці і не повертає нового списку. Таким чином, вираз list(l.pop(-2)).append("тепер це останній елемент") додає рядок "тепер це останній елемент" до списку ['2'], але результат цього виразу - це None, тобто нічого.

знайомство з циклами
for - ітераційний цикл (тобто він повинен пройти задану кількість ітерацій)

while - умовний цикл (ми не знаємо скільки циклів треба виконати, але знаємо умову виходу з циклу)

```python
from random import randint

names = ["Богдан", "Сергій", "Тарас", "Назар", "Василь"]
hobbi = ["футбол", "компютерні ігри", "аніме", "серіали", "хайкінг", "спати"]

for name in names:
    print(f"{name} любить {hobbi[randint(0,5)]}")

print("=|="*10)
# Переписуємо ітераційний цикл за допомогою while
j = 0 #вводимо ітераційну змінну, початкове значення 0 бо ми будемо доступатись до емелентів списку
while j < len(names):
     print(f"{names[j]} любить {hobbi[randint(0,5)]}")
     j+=1

print("=|="*10)

while len(names) > 0:
     print(f"{names.pop()} любить {hobbi[randint(0,5)]}")
```

напишемо програму емуляції кидання кубика (міні казино)
буде задано кількість спроб за які нам потрібно викинути бажане значеннея на кубику

кидати кубик будемо доки не випаде потрібне число або не закінчиться кількість спроб

якщо випало потрібне число в межах кількості спроб то ми виграли

```python
attemps = 3
dice_number = 1 # Значення числа на кубику, від 1 до 6
dice_hit = 0 # початкове значення перед кидання кубика

while dice_hit != dice_number and attemps > 0:
    dice_hit = randint(1,6) # Це буде симулювати кидання кубика, нам буде випадати випадкове число від 1 до 6
    attemps -= 1
    print(f"Випало значення {dice_hit}, воно відповідає бажаному: {dice_hit == dice_number}.")
    if dice_hit == dice_number:
        print("Ви виграли!")
    elif attemps == 0:
        print("Ви програли!")
    else:
        print(f"Спробуйте ще, залишилось {attemps} спроб.")
```

знайомство з функціями та Лямбдами
іменованй кусок коду, який ми можемо викликати щей задаши певні параметри

запитались в ChtaGPT що таке функції, і от маємо пояснення:

У мові програмування "функція" - це блок коду, який виконує певну задачу або виконує певні дії. Функції є однією з основних будівельних блоків програм та дозволяють вам організовувати код у великих програмах на більш менш логічні та повторно використовувані частини.

Основні аспекти функцій:

Назва: Функція має унікальне ім'я, за допомогою якого до неї можна звертатися.

Параметри (аргументи): Функція може приймати вхідні дані, які називаються параметрами або аргументами. Ці параметри використовуються всередині функції для виконання певних операцій.

Код: Функція містить набір інструкцій та операцій, які виконуються при виклику функції. Цей код реалізує конкретний функціонал.

Повернення значення: Функція може повертати значення, як результат її виконання. Це значення може бути використане в інших частинах програми.

Виклик функції: Для виконання функції необхідно її викликати з іншої частини програми, передаючи необхідні аргументи, якщо це потрібно.

Повторне використання: Функції дозволяють вам згрупувати код і використовувати його в інших частинах програми або навіть в інших програмах, забезпечуючи повторне використання коду.

Функції спрощують розробку програм, оскільки дозволяють поділити велику задачу на менші, керовані та більш зрозумілі частини. Вони також полегшують зміну та підтримку коду, оскільки зміни в функціях можуть впливати тільки на їхню внутрішню роботу, не впливаючи на іншу частину програми.

```python
# може просто зробит якісь операції
a = 3
print(f"Сума значення a з 1 буде рівна {a+1}")

# якщо блок коду виконується більше 1 разу, то краще створити для нього функцію
def my_print(a:int):
    print(f"Сума значення a з 1 буде рівна {a+1}")

my_print(4)
```

```python
def check_if_value_is_none(value:str):
    """
    return: Повертає True або False
    """
    _ = value == None
    print(f"Перевіряємо чи значення {value} є рівним None: {_}")
    return _
    print("Це ніколи не виконається бо ми вже вийши з функції!")

check_if_value_is_none("Yes")
check_if_value_is_none("None")
check_if_value_is_none(None)
```

```python
# функції можна викликати або передавати як обєкти

f = check_if_value_is_none
w = check_if_value_is_none("Неважливо")
print(f"Це є НЕ викликана функція яка записано у змінну: {f}\nА це вже результат який повертає викликана функція: {w}")

print(f("Передали у функцію!"))


def pass_fun_inside_fun(fun):
    print("Це код основної функції.")
    fun()
    print("Це вже буде після виклику переданої функції.")

def inside_fun():
    print("Це вивід з функції яку ми передаємо в іншу функцію!")

pass_fun_inside_fun(inside_fun)
```

спробуємо написати Лямбду!

```python
# прототип Лямбди
def lambda_prototype(p:int):
    return [p + i if p > 10 else p - i for i in range(p)]

print(f"Це виклик прототипу Лямбди: {lambda_prototype(11)}")

# а тепер перепишемо це через Лямбду
lm = lambda p: [p + i if p > 10 else p - i for i in range(p)]

print(f"А це вже виклик Лямбди: {lm(9)}")
```
